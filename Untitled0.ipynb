```python
class LSTM(tf.keras.Model):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim):
        super(LSTM, self).__init__()
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        self.lstm = tf.keras.layers.LSTM(hidden_dim, return_sequences=False, return_state=False)
        self.fc = tf.keras.layers.Dense(output_dim)

    def call(self, x):
        out = self.lstm(x)
        out = self.fc(out)
        return out
```

```python
class GRU(tf.keras.Model):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim):
        super(GRU, self).__init__()
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        self.gru = tf.keras.layers.GRU(hidden_dim, return_sequences=False, return_state=False)
        self.fc = tf.keras.layers.Dense(output_dim)

    def call(self, x):
        out = self.gru(x)
        out = self.fc(out)
        return out
```

```python
class RNN(tf.keras.Model):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim):
        super(RNN, self).__init__()
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        self.rnn = tf.keras.layers.SimpleRNN(hidden_dim, return_sequences=False, return_state=False)
        self.fc = tf.keras.layers.Dense(output_dim)

    def call(self, x):
        out = self.rnn(x)
        out = self.fc(out)
        return out
```
